```python
# ============================================================
# STEP 1: USE ALL NUMERIC FEATURES (not just 20)
# ============================================================
ignore_cols = ['Customer_ID', 'churn']

all_num_cols = [
    col for col in data.columns
    if data[col].dtype in ['int64', 'float64'] and col not in ignore_cols
]

print(f"Total numeric features: {len(all_num_cols)}")

X_full = data[all_num_cols].copy()
y = data['churn']

from sklearn.model_selection import train_test_split

X_train_f, X_test_f, y_train, y_test = train_test_split(
    X_full, y, test_size=0.2, random_state=2025, stratify=y
)

# Impute with training medians only
medians = X_train_f.median()
X_train_f = X_train_f.fillna(medians)
X_test_f  = X_test_f.fillna(medians)

print("NaNs remaining:", X_train_f.isna().sum().sum())
```

```python
# ============================================================
# STEP 2: FEATURE ENGINEERING (fixed — runs on full feature set)
# ============================================================
X_train_eng = X_train_f.copy()
X_test_eng  = X_test_f.copy()

# 1. Upgrade Readiness: old device + cheap plan = churn risk
X_train_eng['upgrade_readiness'] = X_train_eng['eqpdays'] / (X_train_eng['hnd_price'] + 1)
X_test_eng['upgrade_readiness']  = X_test_eng['eqpdays']  / (X_test_eng['hnd_price'] + 1)

# 2. Overage Financial Strain (vceovr_Mean IS in the full set now)
X_train_eng['overage_strain'] = X_train_eng['vceovr_Mean'] / (X_train_eng['totmrc_Mean'] + 1)
X_test_eng['overage_strain']  = X_test_eng['vceovr_Mean']  / (X_test_eng['totmrc_Mean'] + 1)

# 3. Usage drop flag
X_train_eng['is_dropping_usage'] = (X_train_eng['change_mou'] < 0).astype(int)
X_test_eng['is_dropping_usage']  = (X_test_eng['change_mou']  < 0).astype(int)

# 4. Revenue drop flag
X_train_eng['is_dropping_rev'] = (X_train_eng['change_rev'] < 0).astype(int)
X_test_eng['is_dropping_rev']  = (X_test_eng['change_rev']  < 0).astype(int)

# 5. Minutes per call (engagement quality)
X_train_eng['mou_per_call'] = X_train_eng['mou_Mean'] / (X_train_eng['attempt_Mean'] + 1)
X_test_eng['mou_per_call']  = X_test_eng['mou_Mean']  / (X_test_eng['attempt_Mean']  + 1)

# 6. Call completion rate (network satisfaction proxy)
X_train_eng['completion_rate'] = X_train_eng['complete_Mean'] / (X_train_eng['attempt_Mean'] + 1)
X_test_eng['completion_rate']  = X_test_eng['complete_Mean']  / (X_test_eng['attempt_Mean']  + 1)

# 7. Revenue per minute (price sensitivity)
# X_train_eng['rev_per_mou'] = X_train_eng['rev_Mean'] / (X_train_eng['mou_Mean'] + 1)
# X_test_eng['rev_per_mou']  = X_test_eng['rev_Mean']  / (X_test_eng['mou_Mean']  + 1)

# 8. Dual drop: both usage AND revenue dropping = high risk
X_train_eng['double_drop'] = (
    (X_train_eng['change_mou'] < 0) & (X_train_eng['change_rev'] < 0)
).astype(int)
X_test_eng['double_drop'] = (
    (X_test_eng['change_mou'] < 0) & (X_test_eng['change_rev']  < 0)
).astype(int)

print(f"Features after engineering: {X_train_eng.shape[1]}")
```

```python
# ============================================================
# STEP 3: CATBOOST WITH BETTER PARAMS 
# ============================================================
from catboost import CatBoostClassifier
from sklearn.metrics import roc_auc_score, classification_report

cat_model = CatBoostClassifier(
    iterations=1000,
    depth=6,
    learning_rate=0.05,
    l2_leaf_reg=5,
    random_seed=2025,
    eval_metric='AUC',
    early_stopping_rounds=50,
    verbose=100
)

cat_model.fit(
    X_train_eng, y_train,
    eval_set=(X_test_eng, y_test)
)

cat_probs = cat_model.predict_proba(X_test_eng)[:, 1]
print(f"\nCatBoost AUC: {roc_auc_score(y_test, cat_probs):.4f}")
#----------------------------------------------------------------------------------------------------------GENAI
cat_preds = (cat_probs >= 0.5).astype(int)

result_data = X_test_eng.copy()
result_data["churn"]        = cat_preds
result_data["churn_prob"]   = cat_probs
result_data["Customer_ID"]  = customer_ids.iloc[X_test_f.index].values

result_data.to_csv("data_with_predictions.csv", index=False)

print(f"Total test rows     : {len(result_data)}")
print(f"High-risk (churn=1) : {result_data['churn'].sum()}")
print(f"\nSample:")
print(result_data[result_data["churn"]==1][["Customer_ID","churn","churn_prob","avgrev","vceovr_Mean"]].head())


```

```python
# ============================================================
# STEP 4: LGBM (usually strongest on this type of data)
# ============================================================
import lightgbm as lgb

# Access it using the prefix
model = lgb.LGBMClassifier()


lgb_model = lgb.LGBMClassifier(
    n_estimators=1000,
    max_depth=6,
    num_leaves=40,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    reg_alpha=0.1,
    reg_lambda=1.0,
    random_state=2025,
    verbose=-1,
    class_weight='balanced'
)

lgb_model.fit(
    X_train_eng, y_train,
    eval_set=[(X_test_eng, y_test)],
    callbacks=[lgb.early_stopping(100), lgb.log_evaluation(150)]
)

lgb_probs = lgb_model.predict_proba(X_test_eng)[:, 1]
print(f"LightGBM AUC: {roc_auc_score(y_test, lgb_probs):.4f}")
```

```python
# ============================================================
# STEP 5: XGBOOST
# ============================================================
from xgboost import XGBClassifier

xgb_model = XGBClassifier(
    n_estimators=1000,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    reg_alpha=0.1,
    reg_lambda=1.0,
    random_state=2025,
    eval_metric='auc',
    early_stopping_rounds=50,
    verbosity=0,
    scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()  
)

xgb_model.fit(
    X_train_eng, y_train,
    eval_set=[(X_test_eng, y_test)],
    verbose=100
)

xgb_probs = xgb_model.predict_proba(X_test_eng)[:, 1]
print(f"XGBoost AUC: {roc_auc_score(y_test, xgb_probs):.4f}")
```

```python
# ============================================================
# STEP 5b: RANDOM FOREST
# ============================================================
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(
    n_estimators=500,
    max_depth=8,
    min_samples_split=5,
    min_samples_leaf=2,
    max_features='sqrt',
    class_weight='balanced',  # helps since churn is ~50/50 but good practice
    random_state=2025,
    n_jobs=-1,
)

rf_model.fit(X_train_eng, y_train)

rf_probs = rf_model.predict_proba(X_test_eng)[:, 1]
print(f"Random Forest AUC: {roc_auc_score(y_test, rf_probs):.4f}")
```

```python
# ============================================================
# STEP 6: UPDATED BLEND (with Random Forest)
# ============================================================

# Print all individual scores first
print("--- Individual Model AUCs ---")
print(f"LightGBM     : {roc_auc_score(y_test, lgb_probs):.4f}")
print(f"CatBoost     : {roc_auc_score(y_test, cat_probs):.4f}")
print(f"XGBoost      : {roc_auc_score(y_test, xgb_probs):.4f}")
print(f"Random Forest: {roc_auc_score(y_test, rf_probs):.4f}")

# Weighted blend — RF gets less weight since it's usually weaker
blended = (
    0.35 * lgb_probs +
    0.30 * cat_probs +
    0.25 * xgb_probs +
    0.10 * rf_probs
)

print(f"\n🏆 Blended Ensemble AUC: {roc_auc_score(y_test, blended):.4f}")

from sklearn.metrics import accuracy_score
preds = (blended >= 0.5).astype(int)
print(f"Accuracy: {accuracy_score(y_test, preds):.4f}")
print(classification_report(y_test, preds))
```

```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. SET UP THE ECONOMIC DATA ---
# Let's assume a pool of 10,000 high-risk subscribers
total_subscribers = 10000
arpu_annual = 480  # $40 per month * 12 months = $480/year per user

# Before Model (Status Quo: 50% churn rate in this high-risk test pool)
revenue_baseline = total_subscribers * arpu_annual
loss_before = (total_subscribers * 0.50) * arpu_annual  # 5,000 users leave
net_profit_before = revenue_baseline - loss_before

# After Model (XGBoost intercepts them, saving 15% of the churners = 750 users saved)
users_saved = 750
incentive_cost = 10000 * 50  # Giving a $50 upgrade voucher to the pool

revenue_after = revenue_baseline
loss_after = loss_before - (users_saved * arpu_annual)  # Loss shrinks!
net_profit_after = revenue_after - loss_after - incentive_cost

# --- 2. PLOT CONFIGURATION (Matching your reference image) ---
labels = ['Before Model', 'After Model']
x = np.arange(len(labels))
width = 0.22

fig, ax = plt.subplots(figsize=(11, 6.5), dpi=100)

# Generate the side-by-side grouped bars
rects1 = ax.bar(x - width, [revenue_baseline/100000, revenue_after/100000], width, label='Gross Contract Value', color='#2F75B5')
rects2 = ax.bar(x, [loss_before/100000, loss_after/100000], width, label='Revenue Lost to Churn', color='#ED7D31')
rects3 = ax.bar(x + width, [net_profit_before/100000, net_profit_after/100000], width, label='Net Retained Revenue', color='#A5A5A5')

# --- 3. STYLING & FORMATTING ---
ax.set_ylabel('Financial Scale (in $100K USD)', fontsize=11, fontweight='bold', labelpad=10)
ax.set_title('Financial Impact Analysis: Proactive Churn Mitigation Framework', fontsize=14, fontweight='bold', pad=20)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=12, fontweight='bold')
ax.set_ylim(0, 58)
ax.grid(axis='y', linestyle='--', alpha=0.5)
ax.legend(loc='upper right', frameon=True, facecolor='white', edgecolor='none')

# Add values on top of the bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'${height:.2f}M' if height >= 10 else f'${height:.2f}K',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=9, fontweight='bold')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

# --- 4. THE EXECUTIVE RED BANNER (The highlight indicator from your reference) ---
# We draw a prominent banner pointing out the exact loss reduction
saved_amount = (loss_before - loss_after) / 100000
banner_text = f"Revenue Loss Reduced by ${saved_amount:.2f}K\n(15% Active Churn Interception)"

ax.text(0.5, 24, banner_text, color='white', weight='bold', ha='center', va='center',
        bbox=dict(boxstyle='rarrow,pad=0.6', facecolor='red', edgecolor='none'))

# Decorative dynamic arrow
ax.annotate('', xy=(0.95, 18), xytext=(0.05, 18),
            arrowprops=dict(arrowstyle="->", color='#2F75B5', lw=2.5))

plt.tight_layout()
plt.show()
```
