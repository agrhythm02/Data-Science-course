```python
from sklearn.linear_model import LinearRegression

# Example data
X = [[1], [2], [3], [4],[5],[7],[9]]
y = [1, 8, 27, 64,125,343,729]

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Predict
print(model.predict([[3]]))  

# Equation
# print(model.coef_)      # slope
# print(model.intercepts_) # intercept
```
