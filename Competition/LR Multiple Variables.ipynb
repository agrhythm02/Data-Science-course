```python
import pandas as pd
import sklearn.linear_model as linear_model
```

```python
df = pd.read_csv('train.csv')
df.describe()
```

```python
# Fill everything with 0
df = df.fillna(0)
```

```python
df.info()
```

```python
reg = linear_model.LinearRegression()
# 1. Define your features (X) and take the first 30 rows
features = ['Id','Year','Age','Height','Weight','Sprint_40yd','Vertical_Jump','Bench_Press_Reps','Broad_Jump','Agility_3cone','Shuttle']
X = df[features].head(5)

# 2. Define your target (y) and take the first 30 rows
y = df.Drafted.head(5)

# 3. Fit the model
reg.fit(X, y)
```
